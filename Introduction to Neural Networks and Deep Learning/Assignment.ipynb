# Part A 

```python
import pandas as pd
import numpy as np
```

```python
concrete_data = pd.read_csv('https://cocl.us/concrete_data')
concrete_data.head()
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(8,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
from sklearn.model_selection import train_test_split
concrete_data_columns = concrete_data.columns
X = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
y = concrete_data['Strength'] # Strength column

scores = []

for epochs in range(0, 50):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    model = regression_model()
    model.fit(X_train, y_train, epochs=50, verbose=0)
    score = model.evaluate(X_test, y_test, verbose=0)
    scores.append(score)
    print("epoch" + str(epochs) + ": " + "Succeed")
print(scores)
```

```python
mean = np.mean(scores)
std = np.std(scores)
print("The mean of the list is {}, the std is {}".format(mean, std))
```

```python
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(8,)))
    model.add(Dense(10, activation='relu', input_shape=(10,)))
    model.add(Dense(10, activation='relu', input_shape=(10,)))
    model.add(Dense(1))

    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

from sklearn.model_selection import train_test_split
concrete_data_columns = concrete_data.columns
X = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
y = concrete_data['Strength'] # Strength column
X = (X - X.mean()) / X.std()
scores = []

for epochs in range(0, 50):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    model = regression_model()
    model.fit(X_train, y_train, epochs=50, verbose=0)
    score = model.evaluate(X_test, y_test, verbose=0)
    scores.append(score)
    print("epoch" + str(epochs) + ": " + str(score))
print(scores)
mean = np.mean(scores)
std = np.std(scores)
print("The mean of the list is {}, the std is {}".format(mean, std))
```

## part B std = 102.7， Part D std = 13.2
## The std of D is smaller than B
